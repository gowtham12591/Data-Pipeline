```python
import pandas as pd

df = pd.read_csv('Data/EmployeeAttrition.csv')
df.head()
```

- Here EmployeeNumber is the primary key, which can be used for mapping the dataset

```python
# Dataset splitting

data1_df = df[['EmployeeNumber', 'Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
               'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount']]
print(data1_df.info())
data1_df.head()
```

```python
data2_df = df.drop(columns=['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
               'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount'], axis=1)
print(data2_df.info())
data2_df.head()
```

```python
data2_df.columns = data2_df.columns.str.lower()
```

```python
# Saving the splitted dataframes as two seperate datasets

data1_df.to_csv('data1.csv', index=False)
data2_df.to_csv('data2.csv', index=False)
```
